In this notebook, we will introduce how to perform calculate transition path for DASH results. 

Please make sure you are running the notebook at the directory of "DASH/tutorial/03_Transition_PATH", because we will use relative path in this notebook. 

---

We have the folder here named "ebpgMD_SplitContrastMAE_CA_algq1_inactive_rt10_nsd20_fcp01000_ecsol_msr0_pct20l10/29", in which stores the physical CVs of accumulated trajectories as "arcv.npy". The "labels_of_combined.npy" in it records the attribute of each frame. This folder is the data for ALGQ1 inactive -> active transition in out article. To make the transition path, let's do MSM reweighting to the free energy surface (FES) first. 

```python
import sys,os
os.chdir("/home/jyzha/project/enhanced_sampling/DASH/tutorial/03_Transition_Path/")
sys.path.append("../../DASH/")
from msm import msm_and_TP


os.chdir("ebpgMD_SplitContrastMAE_CA_algq1_inactive_rt10_nsd20_fcp01000_ecsol_msr0_pct20l10")
msm_and_TP(name="algq1",ni=29,n_bins=20,z_file="arcv.npy")
os.chdir("..")
```

On the FES, there is a very stable inactive state at (83,60) and a metastable active state at (67,55). We can further generate the transition path connecting the two states.


```python
import sys,os
sys.path.append("../../DASH/")
from msm import msm_and_TP


os.chdir("ebpgMD_SplitContrastMAE_CA_algq1_inactive_rt10_nsd20_fcp01000_ecsol_msr0_pct20l10")
msm_and_TP(name="algq1",ni=29,n_bins=20,z_file="arcv.npy",z_st_x=83,z_st_y=60,z_ed_x=67,z_ed_y=55)
os.chdir("..")
```

You can get the transition paths in the same folder.
